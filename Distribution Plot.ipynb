```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
```

```python
xls   = pd.ExcelFile('sales.xls')
sales = pd.read_excel(xls, 'Orders')
sales.head()
```

```python
sns.displot(sales.Sales)
```

```python
sns.displot(sales.Profit)
```

```python
sns.displot(sales.Profit, bins=200)
```

```python
sns.distplot(sales.Profit, kde=False)
```

```python
a = np.random.gamma(6, size=200)
```

```python
sns.displot(a)
```

```python
sns.displot(a, kind='kde', fill=True)
```

```python
sns.displot(a, kind='kde', fill=False)
```

```python
sns.distplot(a, fit=stats.gamma)
```

```python
b = np.random.beta(7,7,size=200)
```

```python
sns.displot(b)
```

```python
sns.distplot(b, kde=False, fit=stats.beta)
```
