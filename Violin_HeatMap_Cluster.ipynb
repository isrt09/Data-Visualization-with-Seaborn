```python
import pandas as pd
import numpy  as np
import seaborn as sns
```

```python
xls   = pd.ExcelFile('sales.xls')
sales = pd.read_excel(xls, 'Orders')
sales.head()
```

```python
sns.catplot(x='Product Category', y="Shipping Cost", data=sales, kind='violin')
```

```python
sns.catplot(x='Product Category', y="Shipping Cost", data=sales, kind='violin', hue='Ship Mode')
```

```python
sns.catplot(x='Product Category', y="Shipping Cost", data=sales, kind='violin', hue='Ship Mode', inner='stick')
```

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
flight.isna().sum()
```

```python
flight = flight.pivot("month", "year", "passengers")
sns.heatmap(flight,linewidths=.5,cmap="YlGnBu")
```

```python
corr = flight.corr()
print(corr)
```

```python
sns.heatmap(corr,linewidths=.5,cmap="YlGnBu")
```

```python
corr = flight.corr()
sns.heatmap(corr,linewidths=.5,cmap="YlGnBu", annot=True)
```
