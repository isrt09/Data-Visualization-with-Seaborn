```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# load Dataset
```

```python
xls   = pd.ExcelFile('sales.xls')
sales = pd.read_excel(xls, 'Orders')
sales.head()
```

```python
# Scatter Plot
```

```python
sns.relplot(x='Sales',y='Profit', data = sales)
```

```python
sns.relplot(x='Sales',y='Profit', data = sales, hue='Order Priority', style='Ship Mode')
```

```python
sns.relplot(x='Sales',y='Profit', data = sales, size='Discount',sizes=(35,200))
```

```python
sns.relplot(x='Order Date', y='Sales', data=sales, kind='line')
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", ci='sd')
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", estimator=None)
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", estimator=sum)
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", hue='Ship Mode')
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", hue='Ship Mode', style='Product Category')
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", hue='Ship Mode', style='Product Category', ci=None)
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", hue='Ship Mode', ci=None, col='Product Category')
```

```python
sns.relplot(x="year",y="Sales", data=sales, kind="line", hue='Ship Mode', ci=None, col='Region', col_wrap=4)
```
