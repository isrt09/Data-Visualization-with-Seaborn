```python
import pandas as pd
import seaborn as sns
```

```python
xls   = pd.ExcelFile('sales.xls')
sales = pd.read_excel(xls, 'Orders')
sales.head(3)
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind='box')
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind='box', hue='Ship Mode')
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind='box', row='Ship Mode')
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind='box', col='Ship Mode')
```

```python
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind='box', col='Ship Mode', row='Customer Segment')
```
