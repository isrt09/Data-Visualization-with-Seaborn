```python
import pandas as pd
import seaborn as sns
```

```python
xls   = pd.ExcelFile('sales.xls')
sales = pd.read_excel(xls, 'Orders')
sales.head(3)
```

```python
sns.relplot(x='Sales', y='Profit', data = sales)
```

```python
sns.relplot(x='Sales', y='Profit', data = sales, col='Customer Segment')
```

```python
sns.relplot(x='Sales', y='Profit', data = sales, col='Customer Segment', col_wrap=2)
```

```python
sns.relplot(x='Sales', y='Profit', data = sales, col='Customer Segment', col_wrap=2, hue='Order Priority')
```

```python
sns.relplot(x='Sales', y='Profit', data = sales, col='Customer Segment', col_wrap=2, hue='Ship Mode')
```

```python
sns.relplot(x='Sales', y='Profit', data = sales, col='Customer Segment', row='Product Category', hue='Ship Mode')
```
